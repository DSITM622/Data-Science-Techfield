```python
# creating dummy variables for columns, droppi
cars = pd.get_dummies(cars)
```

```python
Y = cars['hwy']
type(Y)
```

```python
cars.drop('hwy', axis = 1, inplace = True)
```

```python
cars = (cars - cars.min()) / (cars.max() - cars.min())
```

```python
cars['dum'] = 1
```

```python
def OLS(Y, y_hat):
    return (Y - y_hat).dot(Y - y_hat)

def R2(Y, y_hat):
    return 1 - (Y - y_hat).dot(Y - y_hat)/(Y - Y.mean()).dot(Y - Y.mean())
```

```python
# Gradient Descent
h = 1
w_best = np.random.randn(cars.shape[1])
J_best=OLS(Y,cars.dot(w))

for i in range(h):
    w = np.random.randn(cars.shape[1])
    J = [ ]
    eta = 1.4e-4
    epochs = 6000

    for t in range(epochs):
        y_hat = cars.dot(w)
        J.append(OLS(Y,y_hat))
        w -= eta*cars.T.dot(y_hat - Y)
        
    if J[-1] < J_best:
        J_best=J[-1]
        w_best=w
        print('On Iter {} got R-square {}'.format(i, R2(Y, y_hat)))
    
plt.figure(figsize = (8,6))
plt.plot(J)
print("R squared: {}".format(R2(Y, y_hat)))
```
