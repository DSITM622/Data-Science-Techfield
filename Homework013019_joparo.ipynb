```python
#Set up Packages

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Defining Functions

def sigmud(h):
    return 1 / (1+np.exp(-h))

def accuracy(y, p):
    return np.mean(y == np.round(p))

def Gradient(X, p, y):
    return X.T.dot(p - y)
```

```python
# Simulating Data

D = 2
K = 2 # classes
N = K*1050


X0 = np.random.randn(N//K, D) + np.array((2,2))
X1 = np.random.randn (N//K, D) + np.array((-2,-2))
X = np.vstack((X0,X1))

y = np.array([0] * (N//K) + [1]*(N//K)) # // ensure that it is an integer

plt.figure(figsize = (8,5))
plt.scatter(X[:,0], X[:,1], c = y, alpha = 0.9)   # c = y maps a color to the y label
```

```python
# Data Prep

X = np.hstack((np.ones((N,1)), X))
X.shape
```

```python
# Gradient Descent for Logistic Regression

w = np.random.randn(D+1)

J = [ ]
eta = 1e-3
epochs = 1000

for i in range(epochs):
    p = sigmud(X.dot(w))
    J.append(-y.dot(np.log(p)) - ((1 - y).dot(np.log(1 - p))))
    w -= eta*X.T.dot(p - y)

plt.figure(figsize = (8,6))
plt.plot(J)
```

```python
print('Accuracy {}'.format(accuracy(y,p)))
```
